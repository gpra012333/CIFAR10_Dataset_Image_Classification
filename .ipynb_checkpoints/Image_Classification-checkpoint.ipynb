```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import datasets, models
import tensorflow.keras.layers as tfl
from tensorflow.keras.utils import to_categorical
from tensorflow.math import confusion_matrix
from tensorflow.keras.utils import to_categorical
```

```python
(train_X, train_Y), (test_X, test_Y) = datasets.cifar10.load_data()
```

```python
print(train_X.shape)
print(test_X.shape)
print(train_Y.shape)
print(test_Y.shape)
```

```python
classes = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

def plot_image(X,Y,index):
  plt.figure(figsize=(15,2))
  plt.imshow(X[index])
  plt.xlabel(classes[Y[index][0]])
```

```python
plot_image(train_X, train_Y, 2)
```

```python
plot_image(train_X, train_Y, 26)
```

```python
train_X = train_X/255
test_X = test_X/255

train_Y = to_categorical(train_Y)
test_Y = to_categorical(test_Y)
```

```python
ann = models.Sequential([
    tfl.Flatten(input_shape=(32,32,3)),
    tfl.Dense(3000, activation='relu'),
    tfl.Dense(1000, activation='relu'),
    tfl.Dense(10,activation='sigmoid')
])
```

```python
ann.compile(optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy'])
```

```python
ann.fit(train_X, train_Y, epochs=5)
```

```python
ann.evaluate(test_X,test_Y)
```

```python
pred_test_Y = ann.predict(test_X)
```

```python
conf_mat = confusion_matrix(test_Y.argmax(axis=1), pred_test_Y.argmax(axis=1))
```

```python
import seaborn as sns
plt.figure(figsize=(8,8))
sns.heatmap(conf_mat, annot=True, fmt='.1f', cmap='Blues')
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(test_Y.argmax(axis=1), pred_test_Y.argmax(axis=1) ))
```

```python
cnn = models.Sequential([

        # cnn
        tfl.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(32,32,3)),
        tfl.MaxPooling2D((2,2)),

        tfl.Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
        tfl.MaxPooling2D((2,2)),

        tfl.Conv2D(filters=128, kernel_size=(3,3), activation='relu'),
        tfl.MaxPooling2D((2,2)),

        # Dense
        tfl.Flatten(),
        tfl.Dense(128, activation='relu'),
        tfl.Dense(10, activation='softmax')

    ])
```

```python
cnn.compile(optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy'])
```

```python
cnn.fit(train_X, train_Y, epochs=5)
```

```python
cnn.evaluate(test_X, test_Y)
```

```python
pred_test_Y = cnn.predict(test_X)
```

```python
conf_matrix = confusion_matrix(test_Y.argmax(axis=1), pred_test_Y.argmax(axis=1))
```

```python
print(conf_matrix)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(conf_matrix, annot=True, fmt='.1f', cmap='Blues')
```

```python
print(classification_report(test_Y.argmax(axis=1), pred_test_Y.argmax(axis=1)))
```
